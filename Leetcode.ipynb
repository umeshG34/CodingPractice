```python
#1. Two Sum
# https://leetcode.com/problems/two-sum/

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        
        if len(nums) == 0:
            return false
        
        dict_num ={}
        
        for i in range(len(nums)):           
            try:
                return [i, dict_num[(target-nums[i])]]
            except:
                dict_num[nums[i]] = i
```
